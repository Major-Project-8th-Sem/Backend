```python
df = pd.read_csv('/content/drive/MyDrive/all_data.csv')
```

```python
counts = df['Label'].value_counts().to_dict()
```

```python
df.shape[0] , counts['BENIGN']
```

```python
attacks = df.shape[0] - counts['BENIGN']
nonattacks = counts['BENIGN']
```

```python
attacks * 100 /(attacks+nonattacks)
```

```python
'''
  if attacks no is 30% that means attacks = 30%
                                        x = 70%
                                          x = (7/3)*attacks
'''
```

```python
requirednonattacks = (7/3)*attacks
requirednonattacks
```

```python
totaltoremove = round(nonattacks - requirednonattacks)
totaltoremove
```

# **Removing Strategies ....** 


```python
arr = df.loc[df['Label'] == 'BENIGN']
att = df.loc[df['Label'] != 'BENIGN']
print(type(arr))
```

```python
# shuffling the benign df
df = arr.sample(frac=1).reset_index(drop=True)
df
```

```python
# Using drop() function to delete first n rows
df.drop(index=df.index[:totaltoremove],axis=0, inplace=True)
df.reset_index(inplace=True)
df.head(5)
```
